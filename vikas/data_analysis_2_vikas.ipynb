```python
import pandas as pd
import numpy as np
from glob import glob 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Taking a look at all the data files in the 'Data Cleaning' notebook, we'll go ahead with analyzing df_1, df_3, and df_4 in this notebook. 

```python
df_1 = pd.read_csv('df_1_cleaned.csv')
df_3 = pd.read_csv('df_3_cleaned.csv')
df_4 = pd.read_csv('df_4_cleaned_profits.csv')
```

#Preparing df_1 for analysis

```python
df_1 = df_1.drop(['Unnamed: 0', 'Unnamed: 0.1', 'id'], axis=1)
```

```python
df_1 = df_1.drop(['original_title'], axis=1)
```

```python
df_1['title'] = df_1['title'].str.lower()
```

```python
df_1 = df_1.rename(columns = {'title': 'movie'})
```

```python
df_1
```

```python
df_3
```

```python
# Converting 'primary_title' to lowercase for consistency. Also, dropping original_title to remove duplicate entries.

df_3['primary_title'] = df_3['primary_title'].str.lower()
df_3 = df_3.drop('original_title', axis=1)
df_3
```

```python
df_3 = df_3.rename(columns = {'primary_title': 'movie'})
```

```python
df_3
```

```python
df_4
```

```python
df_4 = df_4.drop('Unnamed: 0', axis=1)
```

```python
df_4
```

```python
#Creating a final Dataframe by joining df_1, df_3, and df_4. 
df_final = df_1.merge(df_3, on='movie').merge(df_4, on='movie')
df_final
```

```python
df_final.to_csv('final_analysis_file.csv')
```

```python
# Dropping non-needed columns from df_final
df_final = df_final.drop(['genre_ids', 'release_date_x', 'id', 'tconst'], axis=1)
df_final
```

```python
df_final = df_final.drop(['start_year', 'runtime_minutes'], axis=1)
```

```python
df_final = df_final.drop_duplicates(subset = 'movie', keep='first')
```

```python
df_final
```

#Adding Month of Release column to try and visualize the relationship of release date and gross revenue.

```python
df_final['month_of_release'] = pd.DatetimeIndex(df_final['release_date_y']).month
df_final
```

```python
df_final['genres'].value_counts()
```

#Creating a new Dataframe monthly_gross to start preparing for the bar graph.

```python
monthly_gross = df_final[['month_of_release', 'worldwide_gross']]
```

```python
monthly_gross = monthly_gross.rename(columns = {'worldwide_gross': 'Worldwide_Gross (In Millions)'})
monthly_gross
```

```python
monthly_gross['Worldwide_Gross (In Millions)'] = monthly_gross['Worldwide_Gross (In Millions)'].map( lambda x: x/1000000)
```

#Creating a new Dataframe to show month-wise average Worldwide Gross. This dataframe will be used for the bar chart.

```python
average_monthly_gross = monthly_gross.groupby('month_of_release', as_index=False).mean()
```

```python
average_monthly_gross
```

```python
df_final.info()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
palette = sns.color_palette("GnBu_d", 2)
ax = sns.lineplot(x= 'month_of_release', y = 'Worldwide_Gross (In Millions)', marker= 'o',palette = palette,  ci=0, markersize = 15, data= average_monthly_gross)
plt.yticks(ax.get_yticks(), ax.get_yticks())
plt.ylabel('Average Movie Gross Revenue (Millions)', fontsize=20, weight='bold')
plt.xlabel('Months', fontsize = 20, weight='bold')
sns.set(style = 'whitegrid', font_scale = 1)
plt.show()
```

```python
df_final.loc[df_final['genres'].str.contains(','), 'genres'] = 'several'
genre_graph_data = df_final.groupby('genres', as_index=False)['worldwide_gross'].mean()
genre_graph_data = genre_graph_data.sort_values(by='worldwide_gross', ascending=True)
genre_graph_data
```

```python
sns.set(font_scale = 1)
fig, ax = plt.subplots()
fig.set_size_inches(40,25)
pal = sns.color_palette('GnBu_d', len(genre_graph_data))
ax = sns.barplot(x='genres', y='worldwide_gross', ci=None, palette=np.array(pal[: :]), data=genre_graph_data)
plt.xlabel('Movie Genres', fontsize=40, weight='bold')
plt.ylabel('Average Gross Revenue for Movies (in Millions of US$)', fontsize=40, weight='bold')
plt.yticks(ax.get_yticks(), ax.get_yticks()/1000000)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.show()
```
